# Newton's method

1. start with an estimate $x_0$
2. $x_{n+1} = x_{n} - H_{f}^{-1}(x_n) \nabla_f(x_n)$

where
- $H_f$: Hessian matrix
- $\nabla_f$ : Gradient

## Quadratic function

$$f(x) = (x_1 + 2x_2 - 7)^2 + (2x_1 + x_2 -5)^2$$

$$\nabla_f = \left[ \begin{array}{c} 2(x_1 + 2x_2 - 7)^2 + 4(2x_1 + x_2 - 5) \\
    4(x_1 + 2x_2 - 7) + 2(2x_1 + x_2 - 5) \end{array} \right]$$

$$H_f = \left[ \begin{array}{cc} 10 &  8\\ 8 & 10 \end{array} \right]$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("science")
```

```python
def f(x):
    return (x[0] + 2*x[1] - 7)**2 + (2*x[0] + x[1] - 5)**2
```

```python
def grad_f(x):
    return np.array([2*(x[0] + 2*x[1] - 7) + 4*(2*x[0] + x[1] - 5), 4*(x[0] + 2*x[1] - 7) + 2*(2*x[0] + x[1] - 5)])
```

```python
hess_f = np.linalg.inv(np.array([10., 8., 8., 10.]).reshape(2,2))
```

```python
xi = np.linspace(-5, 5)
yi = np.linspace(-5, 5)
n = xi.size
zi = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        zi[i,j] = f(np.array([xi[i], yi[i]]))
```

**Contour plot**

```python
fig, ax = plt.subplots(figsize = (3, 3))
ax.contour(xi, yi, zi)
plt.show()
```

**Surface plot**

```python
x_mesh, y_mesh = np.meshgrid(xi, yi)
fig, ax = plt.subplots(subplot_kw = {"projection": "3d"}, figsize = (4, 4))
ax.plot_surface(x_mesh, y_mesh, zi, cmap = plt.cm.coolwarm)
plt.show()
```

**Newton's method**

```python
def newton(x_0: np.array, n_iter = 1000, eps = 1e-3):
    x = x_0.copy()
    for i in range(n_iter):
        x = x_0 - hess_f @ grad_f(x_0)
        if np.linalg.norm(x - x_0) < eps:
            print("Convergence at epoch ", i)
            break
    return x
```

```python
newton(np.array([0., 0.]))
```
