# Quadratic function

$$\min f(x) = \frac{1}{2}x^TAx + b^Tx + c$$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("science")
```

```python
# symmetric positive definite matrix A
A = np.array([.2, .1, .1, .2]).reshape(2,2)
A
```

```python
np.linalg.eigvals(A)
```

```python
b = np.array([.1, .2])
b
```

```python
c = .5
c
```

```python
b.T
```

```python
def f(x, A = A, b = b, c = c):
    return .5 * x.T @ A @ x + b.T @ x + c
```

```python
xi = np.linspace(-10, 10)
yi = np.linspace(-10, 10)
n = xi.size
zi = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        zi[i,j] = f(np.array([xi[i], yi[i]]))
```

**Contour plot**

```python
fig, ax = plt.subplots(figsize = (3, 3))
ax.contour(zi)
plt.show()
```

**Surface plot**

```python
x_mesh, y_mesh = np.meshgrid(xi, yi)
fig, ax = plt.subplots(subplot_kw = {"projection": "3d"}, figsize = (4, 4))
ax.plot_surface(x_mesh, y_mesh, zi, cmap = plt.cm.coolwarm)
plt.show()
```

## Indirect method

$$\nabla f(x) = Ax + b$$
The unique minimizer is
$$x^{*} = -A^{-1}b$$

```python
x_star = -np.linalg.inv(A) @ b
x_star
```

## QR factorization

Factorize the hessian matrix $A = QR$ where $Q$ is an orthogonal matrix ($Q^TQ = I_n$) and $R$ is an upper triangular matrix

```python
Q, R = np.linalg.qr(A)
```

$x^{*}$ is the solution to $Rx = z$ where $z = -Q^Tb$

```python
z = -Q.T @ b
z
```

```python
N = z.size
s = np.zeros(N)
```

```python
for i in reversed(range(N)):
    v = 0.
    for j in range(i + 1, N):
        v += R[i, j] * s[j]
    s[i] = (z[i] - v) / R[i,i]
```

```python
s
```

```python
np.linalg.norm(s - x_star)
```
